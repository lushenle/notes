## 函数

### 函数的定义

- 量和量之间的关系，比如：$A=\pi r^2$
- $y=f(x)$ 其中 $x$ 是自变量，$y$ 是因变量
- 函数在 $x_0$ 处取得的函数值 $y_{0}=y| _{x=x_{0}}=f\left( x_{0}\right)$
- 符号只是一种便是，也可以：$y=g(x)$、$y=\varphi \left( x\right)$、$y=\psi \left( x\right) $

**几种函数：**

- 分段函数：$f\left( x\right) =\begin{cases}\sqrt{x} & x\geq 0\\
-x & x <0\end{cases}$
- 反函数：$h=\dfrac{1}{2}gt^{2}\rightarrow h=h\left( t\right) t=\sqrt{\dfrac{2n}{g}}\rightarrow t=t\left( h\right) $
- 显函数和隐函数：$y=x^{2}+1$, $f(x,y)=0$, $3x+y-4=0$

**几种特性**

- 奇偶性
    - 偶函数：$f(-x)=f(x)$, 关于 $y$ 轴堆成：$f(x)=x^{2}$, $f(-x)=(-x)^{2}=x^{2}=f(x)$
    - 奇函数：$f(-x)=-f(x)$, 关于原点对称：$f(x)=x^{3}$, $f(-x)=(-x)^{3}=-x^{3}=-f(x)$
- 周期性：$f(x+T)=f(x)$
- 单调性：
    - 设 $I$ 是函数 $f(x)$ 的定义域。如果对于任意的 $x_{1},x_{2}\in I$，当 $x_{1} < x_{2}$ 时就有 $f\left( x_{1}\right) \leq f\left( x_{2}\right)$ ，那么就说 $f(x)$ 在 $I$ 上单调递增
    - 如果把 $f\left( x_{1}\right) \leq f\left( x_{2}\right)$ 换成 $f\left( x_{1}\right)  <f\left( x_{2}\right)$ ，那么就说 $f(x)$ 在 $I$ 上严格单调递增

## 极限

### 数列

按照一定次序排列的一列数：$u_{1}, u_{2}, ... u_{n}, ...$, 其中 $u_{n}$ 叫做通项

对于数列 $\left\{ u _{n}\right\}$, 如果当 $n$ 无限增大时，其通项无限接近于一个常数 $A$, 则称该数列以 $A$ 为极限或称数列收敛于 $A$, 否则称数列为发散

$\lim\limits _{n\rightarrow \infty }u_{n}=A$, 或 $u_{n} \rightarrow A (n\rightarrow \infty )$

$\lim\limits _{n\rightarrow \infty }\dfrac{1}{3^{n}}=0,\lim\limits _{n\rightarrow \infty }\dfrac{n}{n+1}=1, \lim\limits _{n\rightarrow \infty }2^{n}$ 不存在

**符号表示**

$x\rightarrow \infty$: 当 $|x|$ 无限增大时

$x\rightarrow +\infty$: 当 $x$ 无限增大时

$x\rightarrow -\infty$: 当 $x$ 无限减小时

$x\rightarrow x_{0}$: 当 $x$ 从 $x_{0}$ 的左右两侧无限接近于 $x_{0}$ 时

$x\rightarrow x_{0}^{+}$: $x$ 从 $x_{0}$ 的右侧无限接近于 $x_{0}$ 时 

$x\rightarrow x_{0}^{-}$: $x$ 从 $x_{0}$ 的左侧无限接近于 $x_{0}$ 时

### 无穷小与无穷大

**无穷小**：以零为极限 $\lim\limits _{x\rightarrow \infty }\dfrac{1}{x}=0$, 则 $\dfrac{1}{x}$ 是 $x\rightarrow \infty$ 时的无穷小；$\lim\limits _{x\rightarrow 2}\left( 3x-6\right) =0$, 则 $3x-6$ 是 $x\rightarrow 2$ 的无穷小

**基本性质**

- 有限个无穷小的代数和扔是无穷小
- 有限个无穷小的积扔是无穷小
- 有界变量与无穷小的积扔是无穷小
- 无限个无穷小之后不一定是无穷小

$\begin{aligned}\lim _{n\rightarrow \infty }\left( \dfrac{1}{n^{2}}+\dfrac{2}{n^{2}}+\ldots +\dfrac{n}{n^{2}}\right)\end{aligned} \\
\begin{aligned}=\lim _{n\rightarrow \infty }\dfrac{\dfrac{n\left( n+1\right) }{2}}{n^{2}}\end{aligned} \\
\begin{aligned}=\lim _{n\rightarrow \infty }\dfrac{n+1}{2n}\end{aligned} \\
\begin{aligned}=\dfrac{1}{2}\end{aligned}$

无穷小的商不一定是无穷小

$\lim\limits _{x\rightarrow 0}\dfrac{x}{2x}=\dfrac{1}{2}$ 

> 计算过程：先约掉 $x$

极限有无穷小的关系：$\lim\limits _{x\rightarrow x_{0}}f\left( x\right) =A$ 的充要条件 $f\left( x\right) =A+\alpha\left( x\right)$, 其中 $\alpha\left( x\right)$ 是 $x\rightarrow x_{0}$ 时的无穷小

**无穷大**: 并不是一个很大的数，是相对于变换过程来说

$\lim\limits _{x\rightarrow x_{0}}f\left( x\right) =\infty$ 或 $f\left( x\right) \rightarrow \infty \left( x\rightarrow x_{0}\right)$

无穷小和无穷大的关系：在自变量的变换的同一过程中，如果 $f(x)$ 为无穷大，那么 $\dfrac{1}{f\left( x\right) }$ 为无穷小

无穷小的比较：$\alpha=\alpha(x), \beta=\beta(x)$ 都是无穷小

$\lim\limits _{x\rightarrow x_{0}}\alpha \left( x\right) =0,\lim\limits _{x\rightarrow x_{0}}\beta \left( x\right) =0$

如果 $\lim\limits _{x\rightarrow x_{0}}\dfrac{\beta }{\alpha }=0$, 则称 $\beta$ 是比 $\alpha$ 高阶无穷小

$\lim\limits _{x\rightarrow x_{0}}\dfrac{\beta }{\alpha }=\infty$, 则称 $\beta$ 是比 $\alpha$ 低阶无穷小

$\lim\limits _{x\rightarrow x_{0}}\dfrac{\beta }{\alpha }=C\neq0$, 则称 $\beta$ 是比 $\alpha$ 同阶无穷小 

**洛必达法则**：

如果函数 $f(x)$ 和 $g(x)$ 满足如下条件:

- $f(x)$ 和 $g(x)$ 都是当 $x\rightarrow a$ 时的无穷小(或无穷大)
- $f(x)$ 和 $g(x)$ 在点 $a$ 的某心邻域内都可导，且 $g'\left( x\right) \neq 0$ 
- $\lim\limits _{x\rightarrow a}\dfrac{f'\left( x\right) }{g'\left( x\right) }$ 存在（或为无穷大）

那么 $\lim\limits _{x\rightarrow a}\dfrac{f\left( x\right) }{g\left( x\right) }=\lim\limits _{x\rightarrow a}\dfrac{f'\left( x\right) }{g'\left( x\right) }$

> 0比0型，无穷比无穷型，都可以用

### 连续性和导数

**函数的连续性**

设函数 $y=f(x)$ 在点 $x$ 的某邻域内有定义，如果当自变量的改变 $\Delta x$ 趋近于 $0$ 时，相应函数的改变 $\Delta y$ 也趋近于 $0$, 则称 $y=f(x)$ 在掉 $x$ 处连续

$\lim\limits _{\Delta x\rightarrow 0}\Delta y=\lim\limits _{\Delta x\rightarrow 0}\left[ f\left( x_{0}+\Delta x\right) -f\left( x_{0}\right) \right] =0$

```python
from sympy.abc import x, y
from sympy import DiracDelta

plot(limit(DiracDelta(y), DiracDelta(x), 0), xlim=(-2, 2), ylim=(-2, 2))
```

**函数 $f(x)$ 在点 $x_{0}$ 处连续，需要满足的条件：**

- 函数在该点处有定义
- 函数在该点处有极限 $\lim\limits _{x\rightarrow x_{0}}f\left( x\right)$ 存在
- 极限值等于函数值 $f(x_0)$

**函数的间断点**

行数 $f(x)$ 在点 $x=x_0$ 处不连续，则称其为函数的间断点

3种情况为间断点：

- 函数 $f(x)$ 在点 $x_0$ 处没有定义
- 极限 $\lim\limits _{x\rightarrow x_{0}}f\left( x\right)$ 不存在
- 满足前两点，但是 $\lim\limits _{x\rightarrow x_{0}}f\left( x\right)\neq f(x)$ 

当 $x\rightarrow x_{0}$ 时，$f(x)$ 的左右极限存在，则称 $x_0$ 为 $f(x)$ 的第一类间断点，否则成为第二类间断点

跳跃间断点：$\lim\limits _{x\rightarrow 0^{-}}f\left( x\right)$ 与 $\lim\limits _{x\rightarrow 0^{+}}f\left( x\right)$ 均存在，但不相等

可去间断点：$\lim\limits _{x\rightarrow 0}f\left( x\right)$ 存在但不等于 $f(x_0)$

函数 $f\left( x\right) =\dfrac{x^{2}-1}{x^{2}-3x+2}$ 的连续性

在点 $x=2$, $x=1$ 处没有定义

$\lim\limits _{x\rightarrow 1^{-}}\dfrac{x^{2}-1}{x^{2}-3x+2}=\lim\limits _{x\rightarrow 1^{-}}\dfrac{x+1}{-x-2}=-2$

$\lim\limits _{x\rightarrow 1^{+}}\dfrac{x^{2}-1}{x^{2}-3x+2}=\lim\limits _{x\rightarrow 1^{+}}\dfrac{x+1}{-x-2}=-2$, 在 $x=1$ 处是可去间断点

$\lim\limits _{x\rightarrow 2^{-}}f\left( x\right) =-\infty$, $\lim\limits _{x\rightarrow 2^{+}}f\left( x\right) =+\infty$, 在 $x=1$ 处是第二类间断点

**导数**

平均速度：$v=\dfrac{s}{t}$ 但是如何表示瞬时速度呢？

瞬时经过的路程：$\Delta s=s\left( t_{0}+\Delta t\right) -s\left( t_{0}\right)$

这一小段的平均速度：$\overline{v}=\dfrac{\Delta S}{0t}=\dfrac{s\left( t_{0}+ot\right) -s\left( t_{0}\right) }{\Delta t}$

当 $\Delta t\rightarrow 0$ 时，也就是瞬时速度了：

$$v\left( t_{0}\right) =\lim\limits _{\Delta t\rightarrow 0}\overline{v}=\lim\limits _{\Delta t\rightarrow 0}\dfrac{\Delta s}{\Delta t}=\lim\limits _{\Delta t\rightarrow 0}\dfrac{s\left( t_{0}+\Delta t\right) -s\left( t_{0}\right) }{\Delta t}$$

如果平均变化率的极限存在，$\lim\limits _{\Delta x\rightarrow 0}\dfrac{\Delta y}{\Delta x}=\lim\limits _{\Delta x\rightarrow 0}\dfrac{f\left( x_{0}+\Delta x\right) -f\left( x_{0}\right) }{\Delta x}$，则称此极限为函数 $y=f(x)$ 在点 $x_0$ 处的导数 $f'\left( x_{0}\right)$

$$y'| _{x=x_{0}},\dfrac{dy}{dx}$$ 或 $$\left.\dfrac{df\left( x\right) }{dx}\right| _{x=x_{0}}$$

**常见导数:**

$(C) '=0$

$\left( x^{\mu }\right) '=\mu \cdot x^{\mu -1}$

$(\sin x) '=\cos x$

$(\cos x) '=-\sin x$

$(\tan x) '=\sec^2x$

$(\cot x) '=-\csc2x$

$(\sec x) '=\sec x \tan x$

$(\csc x) '=-\csc x \cot x$

$(a^x)'=a^x\ln a$

$(e^x)'=e^x$

$\left( \log ax\right) '=\dfrac{1}{x\ln a}$

$(\ln x)'=\dfrac{1}{x}$

$\left( \arcsin x\right) '=\dfrac{1}{\sqrt{1-x^{2}}}$

$\left( \arccos x\right) '=-\dfrac{1}{\sqrt{1-x^{2}}}$

$\left( \arccos x\right) '=\dfrac{1}{\sqrt{1-x^{2}}}$

$\left( arccotx\right) '=-\dfrac{1}{1+x^{2}}$

**运算**

$\left( u\pm v\right) '=u'\pm v'$

$(uv) '=u'v\pm uv'$

$\left( \dfrac{u}{v}\right) '=\dfrac{u'v-uv'}{v^{2}}(u\neq 0)$

$(Cu) '=Cu'$

### 偏导数

对于一元函数 $y=f(x)$ 只存在 $y$ 随 $x$ 的变化： 

二元函数 $z=f(x,y$ 存在 $z$ 随 $x$ 变化的变化率，随 $y$ 变化的变化率，随 $x$、$y$ 同时变化的变化率：



定义：设函数 $z=f(x,y)$ 在点 $(x_0, y_0)$ 的某个邻域内有定义，定 $y=y_0$, 一元函数 $f(x,y_0)$ 在点 $x=x_0$ 处可导，即极限：

$$\lim _{\Delta x\rightarrow 0}\dfrac{f\left( x_{0}+\Delta x,y_{0}\right) -f\left( x_{0},y_{0}\right) }{\Delta x}=A$$

则称 $A$ 为函数 $Z=f(x,y)$ 在点 $(x_0, y_0)$ 处关于自变量 $x$ 的偏导数，记作：

$$f_x(x_0,y_0)$$ 或

$$\left.\dfrac{\partial z}{\partial x}\right\vert_{\begin{aligned}x=x_{0}\\y=y_{0}\end{aligned}}\left.\dfrac{\partial f}{\partial x}\right\vert_{\begin{aligned}x=x_{0}\\y=y_{0}\end{aligned}}\left. Z_x\right\vert _{\begin{aligned}x=x_{0}\\y=y_{0}\end{aligned}}$$

几何意义：$\left.\dfrac{\partial f}{\partial x}\right\vert _{x=x_{0}\\y=y_0}=\dfrac{d}{dx}f\left( x,y_{0}\right)\vert _{x=x_0}$ 是曲线 $\begin{cases}z=f\left( x,y\right) \\
y=y_{0}\end{cases}$ 在点 $M_0$ 处的切线 $M_{0}T_{x}$ 对 $x$ 轴的斜率

几何意义：$\left.\dfrac{\partial f}{\partial y}\right\vert _{x=x_{0}\\y=y_0}=\dfrac{d}{dy}f\left( x_0,y\right)\vert _{y=y_0}$ 是曲线 $\begin{cases}z=f\left( x,y\right) \\
x=x_{0}\end{cases}$ 在点 $M_0$ 处的切线 $M_{0}T_{y}$ 对 $y$ 轴的斜率



**多元函数的偏导数计算方法**

求偏导数时，只要将 $n$ 个自变量中的某一个看成变量，其余的 $n-1$ 个自变量均视为常数，然后按一元函数的求导方法计算即可。

### 梯度

**方向导数**

蚂蚁沿什么方向跑路才能活？



函数：$z=f(x,y)$

$\left| pp'\right| =\rho =\sqrt{\left( \Delta x\right) ^{2}+\left( \Delta y\right) ^{2}}$

$\Delta z=f\left( x+\Delta x,y+\Delta y\right) -f\left( x,y\right) $



如果函数的增量与这两点距离的比例存在，则称此为在 $P$ 点沿着 $L$ 的方向导数

$$\dfrac{\partial f}{\partial l}=\lim _{\rho\rightarrow 0}\dfrac{f\left( x+\Delta x,y+\Delta y\right) -f\left( x,y\right) }{\rho }$$

函数：$f(x,y)$ 在 $x$ 轴正向 $\overrightarrow{e}_{1}=\left\{ 1,0\right\}$, $y$ 轴正向 $\overrightarrow{e}_{2}=\left\{ 0,1\right\}$ 的方向导数分别为：$f_{x},f_{y}$, 负方向导数：$-f_{x},-f_{y}$

定理：如果函数 $z=f(x,y)$ 在点 $P(x,y)$ 是可微分的，那么在该点沿任意方向 $l$ 的方向导数都存在

$$\dfrac{\partial f}{\partial 1}=\dfrac{\partial f}{\partial x}\cos \varphi +\dfrac{\partial f}{\partial y}\sin \varphi$$

$\varphi$ 为 $x$ 轴到 $l$ 的角度



求函数 $z=xe^{2y}$ 在点 $P(1,0)$ 处沿从点 $P(1,0)$ 到点 $Q(2-1)$ 的方向导数

solution: 这里方向 $\overrightarrow{l}$ 即为 $\overrightarrow{PQ}=\left\{1,-1\right\}$, 

故 $x$ 轴到方向 $\overrightarrow{l}$ 的转角 $\varphi=-\dfrac{\pi }{4}$

因为 $\left.\dfrac{\partial z}{\partial x}\right| _{\left( 1,0\right) }= \left.e^{2y}\right| _{\left( 1,0\right) }=1$, $\left.\dfrac{\partial z}{\partial y}\right| \left( 1,0\right) = 2xe^{2y}| _{\left( 1,0\right) }=2$

所求方向导数：

$$\dfrac{\partial z}{\partial l}=\cos \left( -\dfrac{\pi }{4}\right) +2\sin \left( -\dfrac{\pi }{4}\right) =-\dfrac{\sqrt{2}}{2}$$

**梯度**

函数：$z=f(x,y)$ 在平面域内具有连续的一阶偏导数，对于其中每一个点 $P(x,y)$ 都有向量 $\dfrac{\partial f}{\partial x}\overrightarrow{i}+\dfrac{\partial f}{\partial y}\overrightarrow{j}$, 则其称为函数在点 $P$ 的梯度。

$$gradf\left(x,y\right)=\dfrac{\partial f}{\partial x}\overrightarrow{i}+\dfrac{\partial f}{\partial y}\overrightarrow{j}$$

$\overrightarrow{e}=\cos \varphi \overrightarrow{i}+\sin \varphi \overrightarrow{j}$ 是方向 $l$ 上的单位向量

$$\begin{aligned}\dfrac{\partial f}{\partial l}=\dfrac{\partial f}{\partial x}\cos \varphi +\dfrac{\partial f}{\partial y}\sin \varphi =\left\{ \dfrac{\partial t}{\partial x},\dfrac{\partial f}{\partial y}\right\} \cdot \left\{ \cos \varphi ,\sin \varphi\right\} \\
=gradf\left( x,y\right) \cdot \overrightarrow{e}=\left| gradf\left( x,y\right) \right| \cos \theta \end{aligned}$$

$$\theta =\left( gradf\left( x,y\right) ,\overrightarrow{e}\right)$$

只有当 $\cos\left( gradf\left( x,y\right) ,\overrightarrow{e}\right)=1, \dfrac{\partial f}{\partial l}$ 才有最大值

函数在某点的梯度是一个向量，它的方向与方向导数最大值取得的方向一致，其大小正好是最大的方向导数



设 $\overrightarrow{u}=xyz+z^{2}+5$, 求 $gradu$, 并求在点 $M(0,1,-1)$ 处方向导数的最大(小)值

$\begin{aligned}\because \dfrac{\partial u}{\partial x}=yz,\dfrac{\partial u}{0y}=xz,\dfrac{\partial u}{\partial z}=xy+2z\end{aligned},$

$\begin{aligned}  \therefore gradu| _{\left(0,1,-1\right) }=\left(yz,xz,y+2z\right)| _{\left(0,1,-1\right) }=\left(-1,0,-2\right)\end{aligned}$

从而 $\max \left\{  \dfrac{\partial u}{\partial l}| _{m}\right\} =\left\| gradu\right\| =\sqrt{5}$

$\min \left\{  \dfrac{\partial u}{\partial l}| _{m}\right\} =-\left\| gradu\right\| =-\sqrt{5}$

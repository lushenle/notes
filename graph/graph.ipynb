```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
import warnings

%matplotlib inline
```

```python
warnings.filterwarnings(action='ignore')
```

```python
# simple example
g = nx.Graph()
g.add_edge(1, 2)
g.add_edge(1, 3)
nx.draw(g, with_labels=True, node_size=1500, alpha=0.3,
        arrows=False)  # arrows= True or False, default False
plt.show()
```

```python
# DIRECTED

# This time a pair can appear 2 times, in one side or in the other
df = pd.DataFrame({'from': ['D', 'A', 'B', 'C', 'A'],
                  'to': ['A', 'D', 'A', 'E', 'C']})

# that we use the DiGraph function to create the graph
g_directed = nx.from_pandas_edgelist(df, 'from', 'to', create_using=nx.DiGraph())

# Make the graph
nx.draw(g_directed, with_labels=True, node_size=1500, alpha=0.3, arrows=True)
plt.title("Directed")
plt.show()
```

```python
# UNDIRECTED

# this time a pair can appear 2 times
# in one side or in the other
df = pd.DataFrame({'from': ['D', 'A', 'B', 'C', 'A'],
                  'to': ['A', 'D', 'A', 'E', 'C']})

# that we use the Graph function to create the graph
g_undirected = nx.from_pandas_edgelist(df, 'from', 'to', create_using=nx.Graph())

# make the graph
nx.draw(g_undirected, with_labels=True, node_size=1500, alpha=0.3, arrows=False)
plt.title("UN-Directed")
plt.show()
```

```python
import numpy as np
from scipy.special import gamma
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Stylesheets defined in Matplotlib
plt.style.use('fivethirtyeight')
```

```python
# Set up runtime comparisons
n = np.linspace(1, 10, 1000)
labels = ['Constant', 'Logarithmic', 'Linear',
          'Log Linear', 'Quadratic', 'Cubic', 'Exponential', 'Factorial']
big_o = [np.ones(n.shape), np.log(n), n, n *
         np.log(n), n**2, n**3, 2**n, gamma(n)]
```

```python
plt.figure(figsize=(12, 10))
plt.ylim(0, 100)
plt.ylabel('Relative Runtime')
plt.xlabel('Input Size')
for i in range(len(big_o)):
    plt.plot(n, big_o[i], label=labels[i])
    plt.legend(loc=0)

# plt.savefig('big-o-notation')
```
